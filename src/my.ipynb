```python
# Deiniowania zmiennych

x = sp.Symbol('x', real=True)
t = sp.Symbol('t', real=True)
k = sp.Symbol('k', real=True, positive=True)
a = sp.Symbol('a', real=True, positive=True)
h = sp.Symbol('hbar', real=True, positive=True)
m = sp.Symbol('m', real=True, positive=True)
n = sp.Symbol('n', integer=True, nonnegative=True, positive=True)

psi = sp.Function('psi')

x, k, psi(x), h, t
```

```python
eq = sp.Eq(psi(x).diff(x, x), -k**2 * psi(x))
eq
```

```python
psi = sp.dsolve(eq, psi(x), ics={psi(0):0})
psi
```

```python
psi = psi.subs(k, n * sp.pi / a)
psi
```

```python
En = n**2 * sp.pi**2  * h**2/ (2 * m * a**2)
En
```

```python
res = sp.integrate(psi.rhs**2, (x, 0, a))
res
```

```python
C1 = sp.solve(res - 1, sp.Symbol('C1'))[1]
C1
```

```python
psi = psi.subs(sp.Symbol('C1'), C1)
psi
```

```python
Psi = psi.rhs * sp.exp(-sp.I * En * t / h)
Psi
```

```python
Psif = sp.lambdify((x, t, n), Psi.subs({a:1, h:1, m:1}))
```

# Praca domowa

```python
p = lambda f : -sp.I*h*f.diff(x)
```

```python
Psi.conjugate() * p(Psi)
```

```python
p_expected = sp.integrate(Psi.conjugate() * p(Psi), (x, 0, a))
p_expected
```

```python
Psi.conjugate() * p(p(Psi))
```

```python
psquare_expected = sp.integrate(Psi.conjugate() * p(p(Psi)), (x, 0, a))
psquare_expected
```

Zasada nieoznaczoności Heisenberga mówi nam że im lepiej znamy położenie cząstki tym gorzej znamy jej moment a im lepiej znamy jej moment tym gorzej znamy jej położenie $\sigma_p\sigma_x \geq \bar{h}/2$

```python
p_var = psquare_expected - p_expected**2
p_var
```

```python
x_expected = sp.integrate(Psi.conjugate() * x * Psi, (x, 0, a))
xsquare_expected = sp.integrate(Psi.conjugate() * x**2 * Psi, (x, 0, a))
x_var = xsquare_expected - x_expected ** 2
x_var
```

```python
eq = sp.Ge(sp.sqrt(p_var*x_var), h/2)
eq
```

```python
sp.solve(eq.subs({a:1, h:4})) # Nie ważne dla jakiego a lub h wynik jest poprawny bo n > 1 a 3/3.14 < 1
```
